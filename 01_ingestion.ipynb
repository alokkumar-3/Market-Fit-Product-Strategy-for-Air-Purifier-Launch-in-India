```python
import pandas as pd
import os
from sqlalchemy import create_engine

# --- DB credentials ---
db_user = 'root'
db_password = 'Passakr3'
db_host = 'localhost'
db_port = '3306'
db_name = 'airpure'

# --- Create SQLAlchemy Engine ---
engine = create_engine(f"mysql+pymysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")

# --- Folder path ---
csv_folder_path = r'E:\My Docs\DATA learn\Complete projects\AirPure\rpc16_input_files\Dataful_Datasets\Cleaned_data'

# --- Ingestion Function ---
def ingest_db(file_path, table_name, engine):
    try:
        print(f" Loading {table_name}...")
        df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')

        # Write to SQL safely 
        df.to_sql(
            table_name,
            con=engine,
            if_exists='replace',
            index=False,
            chunksize=10000
        )

        print(f" Successfully ingested {table_name} ({len(df)} rows)")

    except Exception as e:
        print(f" Error ingesting {table_name}: {e}")

# --- Loop through all CSV files ---
for file in os.listdir(csv_folder_path):
    if file.endswith('.csv'):
        table_name = file[:-4]  # strip ".csv"
        file_path = os.path.join(csv_folder_path, file)
        ingest_db(file_path, table_name, engine)
```
