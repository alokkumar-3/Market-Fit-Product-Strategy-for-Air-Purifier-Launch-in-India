# **Objective**
**The objective of this step is to translate pollution severity and demand-trigger insights into clear product decisions. Rather than designing a single generic air purifier, the strategy focuses on aligning product complexity and feature sets with different city-level demand behaviors.**

Based on the analysis, cities were grouped into three demand buckets — Immediate Launch, Core Market, and Emerging/Watchlist. Each bucket represents a distinct combination of pollution severity, behavioral triggers, and willingness to pay.

To avoid over-fragmentation while still addressing local needs, AirPure Innovations should adopt a three-product archetype strategy, where each archetype is mapped to a demand bucket rather than individual cities.

# **Product Archetypes**

# **Product 1: Shield (Premium) — Immediate Launch Cities**

Immediate Launch cities experience chronic air pollution combined with frequent panic-trigger events such as clustered severe AQI days or high AQI volatility. Consumers in these cities exhibit urgency-driven purchase behavior and higher willingness to pay for perceived health protection.

**Features:-**

True HEPA H13 filtration (PM2.5-first)

High CADR for rapid air turnover

Real-time indoor and outdoor AQI display

Automatic mode triggered by AQI spikes

Mobile alerts during severe pollution events

# **Product 2: Base (Volume) — Core Market Cities**

Core Market cities experience persistently poor air quality, but pollution has become normalized rather than panic-inducing. Demand in these cities is steady and price-sensitive, driven by long-term exposure rather than acute spikes.

**Features:-**

HEPA H12/H13 filtration

Basic activated carbon layer

Large room coverage

Extended filter life to reduce replacement costs

**Base is positioned as reliable daily protection for polluted urban environments.**

# **Product 3: Entry (Introductory) — Emerging / Watchlist Cities**

Emerging cities show early warning signals such as rising AQI volatility or episodic severe pollution, but have not yet reached chronic severity levels. Demand here is awareness-driven and cautious.

**Features:-**

Compact HEPA filtration

Simple AQI indicator (LED-based)

Portable, room-specific usage

**Entry acts as a market seeding product, enabling first-time adoption and creating an upgrade pathway as pollution severity and awareness increase.**
